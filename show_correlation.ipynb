```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
import progressbar as pb

gridsize = 0.01
maxval = 1
minval = -1

df = pd.read_csv('predictions_close.csv')

n = math.ceil((maxval - minval)/gridsize)

grid = np.zeros((n, n))

bar = pb.ProgressBar(n*n)
bar.start()
# symbol = 'AUD_USD_close'
for k, row in df.iterrows():
    # if symbol in row['estimator']:
    i = int(min(max((row['prediction']-minval)/gridsize, 0), n-1))
    j = int(min(max((row['actual']-minval)/gridsize, 0), n-1))
    # print(i, j)
    grid[i, j] += 1
    bar.update(k)

# grid = grid / np.max(np.max(grid))
grid = np.log(grid)

plt.figure()
plt.imshow(grid)
plt.show()
```
