```python
import dataset
import pandas as pd
import datetime

def tofloat(innumber):
    try:
        return float(innumber)
    except:
        return 0

candle_path =  'sqlite:///../../data/barsave.db'

db = dataset.connect(candle_path)

df = []
# make a dataframe
for row in db.query('select * from dailycandles;'):
    row['date'] = datetime.datetime.strptime(row['date'],'%Y-%m-%d')
    for flkey in ['open','close','high','low','volume']:
        row[flkey] = tofloat(row[flkey])
    df.append(row)

df = pd.DataFrame(df)
```
