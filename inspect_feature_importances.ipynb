```python
import pandas as pd
import numpy as np
from matplotlib import pyplot

df = pd.read_csv('importance.csv')

features= np.unique(df['label'])

max_importances = []

for feature in features:
    maxval = df[df['label'] == feature]['importance'].mean()
    max_importances.append({'feature': feature, 'importance': maxval })
max_importances = sorted(max_importances, key = lambda x: x.get('importance'), reverse=True)

for it in max_importances:
    print(it['feature'] + ' ' + str(it['importance']))
```

```python
mean_gain = df['importance'].mean()

knockout_threshold = mean_gain / 2
knockstr = ''
for it in max_importances:
    if it['importance'] <= mean_gain:
        knockstr += "'" + it['feature'] + "',"
print(knockstr)
```
